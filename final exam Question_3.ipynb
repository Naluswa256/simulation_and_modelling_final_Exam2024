```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics


```

```python
dataset = pd.read_csv('./Risk_Factors.csv')

# Verify the imported data
dataset.head()
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Risk_Factor'] = le.fit_transform(dataset['Risk_Factor'])
```

```python
feature_cols = ['Year', 'Fatal', 'Serious', 'Minor','Total_Crashes']
X = dataset[feature_cols]
y = dataset['Risk_Factor']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
expected = y_test
#Predict the response for test dataset
predicted = clf.predict(X_test)
```

```python
print(metrics.confusion_matrix(expected, predicted))
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score, jaccard_score

# Calculate the recall
recall = recall_score(expected, predicted, average='macro')

# Calculate the precision
precision = precision_score(expected, predicted,average='macro')

# Calculate the f1-score
f1 = f1_score(expected, predicted,average='macro')

# Calculate the accuracy
accuracy = accuracy_score(expected, predicted)

# Calculate the Jaccard index
jaccard = jaccard_score(expected, predicted,average='macro')

# Print the results
print("Recall:", recall)
print("Precision:", precision)
print("F1-Score:", f1)
print("Accuracy:", accuracy)
print("Jaccard Index:", jaccard)
```
